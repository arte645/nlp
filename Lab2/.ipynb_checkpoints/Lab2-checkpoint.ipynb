## Предобработка

```python
import re

def clean_text(text):
    text = text.lower()
    
    text = re.sub(r'[^а-яА-ЯёЁa-zA-Z\s]', '', text)
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df["public_petition_text"] = df["public_petition_text"].transform(clean_text)
```

## Токенизация

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from pymorphy3 import MorphAnalyzer
from nltk.corpus import stopwords
```

```python
morph = MorphAnalyzer()
stop_words = stopwords.words('russian')

documents = [
    [
        word_norm for word in word_tokenize(doc, language='russian')
        if (word_norm := morph.normal_forms(word)[0]) not in stop_words
    ] for doc in df["public_petition_text"]
]
documents
```

В тексте попадаются какие то ххх, избавимся от них, от маленьких слов и от слов, написанных с ошибками

```python
from collections import Counter

def preprocess_data(data):
    all_words = []
    for sublist in data:
        all_words.extend(sublist)

    word_count = Counter(all_words)

    for i, sublist in enumerate(data):
        
        filtered_words = [word for word in sublist if len(word) >= 2 and 'хх' not in word and word_count[word] >= 2]
        data[i] = filtered_words

    return data

processed_data = preprocess_data(documents)
print(processed_data)
```

```python
processed_data
```

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences=processed_data, vector_size=512, window=3, min_count=1, workers=4)

embedding = model.wv['снег']
```

```python
embedding 
```

```python
similar_words = model.wv.most_similar("снег", topn=3)
print("Слова, похожие на 'снег':", similar_words)
```

## PCA

```python
import numpy as np
words = list(set(word for sentence in processed_data for word in sentence))
embedding_matrix = np.zeros((len(words), model.vector_size))
embedding_matrix.shape
```

```python
for i, word in enumerate(words):
    embedding_matrix[i] = model.wv[word]
```

```python
from sklearn.decomposition import PCA

n_pca = PCA(n_components=100).fit(embedding_matrix)
print(f'[INFO] {n} components save {sum(n_pca.explained_variance_ratio_):.3%} information.')
np.savetxt(f'ncomp={n}.tsv', n_pca.transform(embedding_matrix), delimiter='\t')
```

```python
vectDf = pd.DataFrame(data=n_pca.transform(embedding_matrix))
```

## Реализация CBoW

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import Counter
from nltk.tokenize import word_tokenize

tokenized_corpus = processed_data[:1000]

# Создание словаря
word_freq = Counter(word for sentence in tokenized_corpus for word in sentence)
vocab = {word: i for i, (word, _) in enumerate(word_freq.items())}
inv_vocab = {i: word for word, i in vocab.items()}

# Параметры
embedding_dim = 10
context_size = 2  # Количество слов вокруг целевого слова
total_context_size = context_size * 2  # Итого размеры контекста

# Подготовка данных
def create_training_data(tokenized_corpus, vocab, context_size):
    X, y = [], []
    for sentence in tokenized_corpus:
        for i, word in enumerate(sentence):
            # Сформируем контекстные слова
            start = max(0, i - context_size)
            end = min(len(sentence), i + context_size + 1)
            context_words = [sentence[j] for j in range(start, end) if j != i]

            # Получаем индексы контекстных слов
            context_indices = [vocab[w] for w in context_words]

            # Обеспечиваем фиксированный размер контекста
            if len(context_indices) < total_context_size:
                context_indices += [0] * (total_context_size - len(context_indices))  # Используем 0 для паддинга

            X.append(context_indices)
            y.append(vocab[word])  # Получаем индекс целевого слова

    return torch.tensor(X, dtype=torch.long), torch.tensor(y, dtype=torch.long)

# Создаем тренировочные данные
X, y = create_training_data(tokenized_corpus, vocab, context_size)

# Проверка: вывод размеров
print(f"Размер словаря: {len(vocab)}")
print(f"Размеры X: {X.shape}, размеров y: {y.shape}")  # Проверяем размеры входных и целевых данных

# CBOW модель
class CBOW(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(CBOW, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.linear = nn.Linear(embedding_dim, vocab_size)

    def forward(self, input):
        embedded = self.embeddings(input)
        embedded = self.linear(embedded)
        return embedded.mean(dim=1)

# Обучение модели
model = CBOW(len(vocab), embedding_dim)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.5)

# Обучение
num_epochs = 1000
for epoch in range(num_epochs):
    total_loss = 0
    model.train()
    optimizer.zero_grad()

    output = model(X)
    loss = loss_fn(output, y)
    loss.backward()
    optimizer.step()

    total_loss += loss.item()
    
    if (epoch + 1) % 10 == 0:
        print(f'Epoch: {epoch + 1}, Loss: {total_loss}')

def get_word_vector(word):
    word_idx = vocab[word]
    return model.embeddings(torch.tensor([word_idx]))
```
